```python
import os
from pydub import AudioSegment
from tqdm import tqdm

# 데이터셋 경로 설정
dataset_path = "/content/drive/MyDrive/프로젝트/케어크루즈 인턴/data/maestro_v3"

# 전처리된 파일을 저장할 새로운 폴더 생성
processed_folder = os.path.join(dataset_path, "finetuning_data")
os.makedirs(processed_folder, exist_ok=True)

# 이미 처리된 파일 목록 수집 (전체 경로 포함)
processed_files = set(
    os.path.join(processed_folder, f) for f in os.listdir(processed_folder) if f.endswith(".wav")
)

# 처리되지 않은 파일 목록 저장
unprocessed_files = []

# 모든 하위 폴더의 파일을 처리
file_list = []
for root, dirs, files in os.walk(dataset_path):
    for filename in files:
        # ._로 시작하는 파일 무시
        if filename.startswith("._"):
            continue
        if filename.endswith(('.mp3', '.wav', '.flac')):
            file_list.append(os.path.join(root, filename))

# tqdm으로 진행 상태 표시
for original_path in tqdm(file_list, desc="Processing Audio Files", unit="file"):
    try:
        # 파일명 확인
        filename = os.path.basename(original_path)
        processed_filename = f"{os.path.splitext(filename)[0]}.wav"
        processed_path = os.path.join(processed_folder, processed_filename)

        # 이미 처리된 경우 건너뛰기
        if processed_path in processed_files:
            print(f"Skipping already processed file: {processed_path}")
            continue

        # 오디오 파일 로드
        try:
            audio = AudioSegment.from_file(original_path)
        except Exception as load_error:
            raise RuntimeError(f"Error loading file: {load_error}")

        # 샘플링 속도를 44100Hz로 변환
        try:
            audio = audio.set_frame_rate(44100)
        except Exception as resample_error:
            raise RuntimeError(f"Error resampling file: {resample_error}")

        # 초반 30초(30,000ms) 추출
        first_30_seconds = audio[:30000]

        # 저장 경로 설정
        try:
            first_30_seconds.export(processed_path, format="wav")
        except Exception as export_error:
            raise RuntimeError(f"Error exporting file: {export_error}")

        # 처리된 파일 기록에 추가
        processed_files.add(processed_path)
    except Exception as e:
        # 처리 실패 시 파일명을 기록
        print(f"Error processing {original_path}: {e}")
        unprocessed_files.append(original_path)

# 처리되지 않은 파일 출력
if unprocessed_files:
    print("\nThe following files could not be processed:")
    for unprocessed_file in unprocessed_files:
        print(unprocessed_file)
else:
    print("\nAll files were processed successfully.")

print(f"Processed files are saved in: {processed_folder}")
```

```python
from essentia.standard import MonoLoader, TensorflowPredict2D
import numpy as np

def filter_predictions(predictions, class_list, threshold=0.1):
    predictions_mean = np.mean(predictions, axis=0)
    sorted_indices = np.argsort(predictions_mean)[::-1]
    filtered_indices = [i for i in sorted_indices if predictions_mean[i] > threshold]
    filtered_labels = [class_list[i] for i in filtered_indices]
    filtered_values = [predictions_mean[i] for i in filtered_indices]
    return filtered_labels, filtered_values

def make_comma_separated_unique(tags):
    seen_tags = set()
    result = []
    for tag in ', '.join(tags).split(', '):
        if tag not in seen_tags:
            result.append(tag)
            seen_tags.add(tag)
    return ', '.join(result)

def get_audio_features(audio_filename):
    audio = MonoLoader(filename=audio_filename, sampleRate=16000, resampleQuality=4)()

    result_dict = {}

    # Predict genres
    genre_model = TensorflowPredict2D(
        graphFilename="/content/drive/MyDrive/프로젝트/케어크루즈 인턴/essentia/genre_discogs400-discogs-effnet-1.pb",
        input="serving_default_model_Placeholder",
        output="PartitionedCall:0"
    )
    predictions = genre_model(audio)
    filtered_labels, _ = filter_predictions(predictions, genre_labels)
    result_dict['genres'] = make_comma_separated_unique(filtered_labels)

    # Predict mood/theme
    mood_model = TensorflowPredict2D(
        graphFilename="/content/drive/MyDrive/프로젝트/케어크루즈 인턴/essentia/mtg_jamendo_moodtheme-discogs-effnet-1.pb"
    )
    predictions = mood_model(audio)
    filtered_labels, _ = filter_predictions(predictions, mood_theme_classes, threshold=0.05)
    result_dict['moods'] = make_comma_separated_unique(filtered_labels)

    return result_dict
```

```python
from essentia.standard import MonoLoader, TensorflowPredict2D
import numpy as np

def filter_predictions(predictions, class_list, threshold=0.1):
    predictions_mean = np.mean(predictions, axis=0)
    sorted_indices = np.argsort(predictions_mean)[::-1]
    filtered_indices = [i for i in sorted_indices if predictions_mean[i] > threshold]
    filtered_labels = [class_list[i] for i in filtered_indices]
    filtered_values = [predictions_mean[i] for i in filtered_indices]
    return filtered_labels, filtered_values

def make_comma_separated_unique(tags):
    seen_tags = set()
    result = []
    for tag in ', '.join(tags).split(', '):
        if tag not in seen_tags:
            result.append(tag)
            seen_tags.add(tag)
    return ', '.join(result)

def adjust_audio_length(audio, target_length=1280):
    """
    오디오 데이터의 길이를 모델의 기대값에 맞게 조정합니다.
    """
    if len(audio) > target_length:
        return audio[:target_length]  # 초과 부분 자르기
    else:
        return np.pad(audio, (0, target_length - len(audio)), 'constant')  # 부족한 부분 0으로 패딩

def prepare_input_for_model(audio):
    """
    모델 입력에 적합한 형식으로 오디오 데이터를 변환합니다.
    """
    if audio.ndim == 1:  # 1D 배열인 경우
        audio = np.expand_dims(audio, axis=0)  # 2D 배열로 변환
    return audio

def get_audio_features(audio_filename):
    audio = MonoLoader(filename=audio_filename, sampleRate=16000, resampleQuality=4)()

    # 입력 크기 조정 및 형식 확인
    adjusted_audio = adjust_audio_length(audio)
    prepared_audio = prepare_input_for_model(adjusted_audio)

    result_dict = {}

    # predict genres
    genre_model = TensorflowPredict2D(graphFilename="/content/drive/MyDrive/프로젝트/케어크루즈 인턴/essentia/genre_discogs400-discogs-effnet-1.pb",
                                      input="serving_default_model_Placeholder", output="PartitionedCall:0")
    predictions = genre_model(prepared_audio)
    filtered_labels, _ = filter_predictions(predictions, genre_labels)
    filtered_labels = ', '.join(filtered_labels).replace("---", ", ").split(', ')
    result_dict['genres'] = make_comma_separated_unique(filtered_labels)

    # predict mood/theme
    mood_model = TensorflowPredict2D(graphFilename="/content/drive/MyDrive/프로젝트/케어크루즈 인턴/essentia/mtg_jamendo_moodtheme-discogs-effnet-1.pb")
    predictions = mood_model(prepared_audio)
    filtered_labels, _ = filter_predictions(predictions, mood_theme_classes, threshold=0.05)
    result_dict['moods'] = make_comma_separated_unique(filtered_labels)

    return result_dict

import os
import json
import random
import librosa
from tqdm import tqdm

# JSON 메타데이터 파일 경로
metadata_path = "/content/drive/MyDrive/프로젝트/케어크루즈 인턴/data/maestro_v3/musicgen_metadata_with_description.json"

# 데이터셋 경로
dataset_path = "/content/drive/MyDrive/프로젝트/케어크루즈 인턴/data/maestro_v3/finetuning_data"

# 메타데이터 로드
with open(metadata_path, "r") as f:
    metadata = json.load(f)

# 오디오 파일 이름만 추출하여 메타데이터 매핑
metadata_dict = {os.path.basename(entry['audio']): entry['metadata'] for entry in metadata}

# Train/Eval 데이터 저장 폴더 생성
output_dir = "/content/drive/MyDrive/프로젝트/케어크루즈 인턴/data/maestro_v3/egs"
os.makedirs(f"{output_dir}/train", exist_ok=True)
os.makedirs(f"{output_dir}/eval", exist_ok=True)

train_len = 0
eval_len = 0
files_without_metadata = []

with open(f"{output_dir}/train/data.jsonl", "w") as train_file, \
     open(f"{output_dir}/eval/data.jsonl", "w") as eval_file:

    # 데이터셋 파일 목록 생성
    dataset_files = [
        os.path.join(root, file)
        for root, _, files in os.walk(dataset_path)
        for file in files if file.endswith(('.wav', '.mp3', '.flac'))
    ]

    # 파일 목록을 섞음
    random.shuffle(dataset_files)

    # tqdm으로 진행 상태 표시
    dset = tqdm(dataset_files, desc="Processing Audio Files", unit="file")

    for filepath in dset:
        filename = os.path.basename(filepath)

        # 메타데이터 가져오기
        metadata_entry = metadata_dict.get(filename)
        if not metadata_entry:
            files_without_metadata.append(filename)
            print(f"메타데이터가 없는 파일: {filename}, 건너뜀")
            continue

        try:
            # 오디오 분석
            y, sr = librosa.load(filepath, sr=44100)
            tempo, _ = librosa.beat.beat_track(y=y, sr=sr)

            # numpy.ndarray 처리
            if isinstance(tempo, np.ndarray):
                tempo = int(np.round(tempo[0]))  # numpy 배열의 첫 번째 값을 반올림 후 int로 변환
            else:
                tempo = int(round(tempo))  # 일반 float 값 처리

            chroma = librosa.feature.chroma_stft(y=y, sr=sr)
            key = np.argmax(np.sum(chroma, axis=1))
            key = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B'][key]
            length = librosa.get_duration(y=y, sr=sr)

            # get_audio_features 호출
            audio_features = get_audio_features(filepath)

            # JSONL에 저장할 항목 생성
            entry = {
                "key": key,
                "artist": metadata_entry.get("composer", ""),
                "sample_rate": 44100,
                "file_extension": "wav",
                "description": metadata_entry.get("description", ""),
                "keywords": metadata_entry.get("top_tags", ""),
                "title": metadata_entry.get("title", ""),
                "name": metadata_entry.get("composer", ""),
                "duration": length,
                "bpm": tempo,
                "genre": audio_features.get('genres', ""),
                "moods": audio_features.get('moods', ""),
                "path": os.path.join(dataset_path, filename),
            }

            # Train/Test 데이터 분할
            if random.random() < 0.85:
                train_len += 1
                train_file.write(json.dumps(entry) + '\n')
            else:
                eval_len += 1
                eval_file.write(json.dumps(entry) + '\n')

        except Exception as e:
            # 에러 발생 시 파일명과 경로 출력
            print(f"오류 발생: 파일: {filename}, 경로: {filepath}, 에러: {e}")
            continue

# 메타데이터 없는 파일 출력
if files_without_metadata:
    print("\n메타데이터가 없는 파일 목록:")
    for file in files_without_metadata:
        print(file)

# 처리 결과 출력
print(f"Train 샘플: {train_len}, Eval 샘플: {eval_len}")
```

```python
import json

# 파일 경로 설정
eval_file_path = "/content/drive/MyDrive/프로젝트/케어크루즈 인턴/data/maestro_v3/egs/eval/data.jsonl"
train_file_path = "/content/drive/MyDrive/프로젝트/케어크루즈 인턴/data/maestro_v3/egs/train/data.jsonl"

# 필드에서 최대 n개의 항목만 추출
def trim_field_values(field, max_items=3):
    if isinstance(field, str):
        return ', '.join(field.split(", ")[:max_items])
    return field

# JSONL 파일 처리 함수
def process_jsonl_file(file_path, output_file_path, max_items=3):
    processed_lines = []
    with open(file_path, "r") as infile:
        for line in infile:
            entry = json.loads(line)

            # genre와 moods 필드를 최대 3개로 제한
            entry['genre'] = trim_field_values(entry.get('genre', ""), max_items)
            entry['moods'] = trim_field_values(entry.get('moods', ""), max_items)

            processed_lines.append(entry)

    # 결과 저장
    with open(output_file_path, "w") as outfile:
        for entry in processed_lines:
            outfile.write(json.dumps(entry) + '\n')
    print(f"Processed file saved to: {output_file_path}")

# eval과 train 파일 각각 처리
process_jsonl_file(eval_file_path, eval_file_path, max_items=3)
process_jsonl_file(train_file_path, train_file_path, max_items=3)
```
